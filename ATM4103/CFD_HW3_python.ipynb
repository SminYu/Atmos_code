```python
import matplotlib.pyplot as plt
```

```python
def k_value(t,x,y,z):
    k = 10*(y-x)
    return k

def l_value(t,x,y,z):
    l = 28*x - y - x*z
    return l

def m_value(t,x,y,z):
    m = x*y - 8*z/3
    return m
```

```python
t_time = 2000
h = 20/t_time
```

```python
#Initiation
li_t = [0]

li_x1 = [5]
li_y1 = [5]
li_z1 = [5]

li_x2 = [5.001]
li_y2 = [5.001]
li_z2 = [5.001]

# RK2 method
for i in range(t_time):
    x = li_x1[i]
    y = li_y1[i]
    z = li_z1[i]    
    t = li_t[i]
    
    k1 = k_value(t,x,y,z)
    l1 = l_value(t,x,y,z)
    m1 = m_value(t,x,y,z)

    k2 = k_value(t+h, x+h*k1, y+h*l1, z+h*m1)
    l2 = l_value(t+h, x+h*k1, y+h*l1, z+h*m1)
    m2 = m_value(t+h, x+h*k1, y+h*l1, z+h*m1)
    
    li_x1.append(x + h*(k1+k2)/2)
    li_y1.append(y + h*(l1+l2)/2)
    li_z1.append(z + h*(m1+m2)/2)
    
    x = li_x2[i]
    y = li_y2[i]
    z = li_z2[i]    
    t = li_t[i]
    
    k1 = k_value(t,x,y,z)
    l1 = l_value(t,x,y,z)
    m1 = m_value(t,x,y,z)

    k2 = k_value(t+h, x+h*k1, y+h*l1, z+h*m1)
    l2 = l_value(t+h, x+h*k1, y+h*l1, z+h*m1)
    m2 = m_value(t+h, x+h*k1, y+h*l1, z+h*m1)
    
    li_x2.append(x + h*(k1+k2)/2)
    li_y2.append(y + h*(l1+l2)/2)
    li_z2.append(z + h*(m1+m2)/2)
    
    t += dt
    li_t.append(t)
```

```python
plt.figure(figsize=(8,6), facecolor = 'w')
plt.plot(li_t, li_x1, 'r', label='5')
plt.plot(li_t, li_x2, 'b', label='5.001')
plt.legend()
plt.title('t-x plot (RK2 method)')
plt.xlabel('t') ; plt.ylabel('x')
plt.show()

plt.figure(figsize=(8,6), facecolor = 'w')
plt.plot(li_t, li_y1, 'r', label='5')
plt.plot(li_t, li_y2, 'b', label='5.001')
plt.legend()
plt.title('t-y plot (RK2 method)')
plt.xlabel('t') ; plt.ylabel('y')
plt.show()

plt.figure(figsize=(8,6), facecolor = 'w')
plt.plot(li_t, li_z1, 'r', label='5')
plt.plot(li_t, li_z2, 'b', label='5.001')
plt.legend()
plt.title('t-z plot (RK2 method)')
plt.xlabel('t') ; plt.ylabel('z')
plt.show()
```
