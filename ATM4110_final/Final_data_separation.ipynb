```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
path = '/home/e2018136018/notebook/data/'

# csv와 xlsx형식의 모든 자료
all_files_csv = glob.glob(path + '*.csv')
all_files_excel = glob.glob(path + '*.xlsx')
li = [] #여기에 파일이 하나씩 쌓임

for filename in all_files_csv:
    df = pd.read_csv(filename, encoding='CP949')
    print(filename)
    li.append(df)

for filename in all_files_excel:
    df = pd.read_excel(filename)
    print(filename)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True) #쌓인 파일을 합쳐서 하나로 만듦.
```
