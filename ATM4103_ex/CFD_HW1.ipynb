```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
t_start = 0 ; y_start = 0
t_end = 10
li_nt = [10,20,40,80,160]

last_err_for = [] ; last_err_back = []

for i, nt in zip(list(range(6)), li_nt):
    h = (t_end-t_start)/nt
    li_t = [t_start]
    li_real_y = [y_start] ; li_for_y = [y_start] ; li_back_y = [y_start]
    li_err_for = [0] ; li_err_back = [0]
    
    t = t_start ; for_y = y_start ; back_y = y_start
    while (t < t_end):
        # t = t
        for_y = h*(for_y+1)*np.cos(t)+for_y
        li_for_y.append(for_y)
        
        # t = t+h
        t += h
        li_t.append(t)
        
        back_y = (h*np.cos(t)+back_y)/(1-h*np.cos(t))
        li_back_y.append(back_y)
        
        real_y = -1. + np.exp(np.sin(t))
        li_real_y.append(real_y)
        
        li_err_for.append(real_y - for_y)
        li_err_back.append(real_y - back_y)
        
    last_err_for.append('%5.3f' % (real_y - for_y))
    last_err_back.append('%5.3f' % (real_y - back_y))

    plt.figure(figsize=(8,6), facecolor='w')
    plt.plot(li_t, li_for_y, label='Forward', c='r')
    plt.plot(li_t, li_back_y, label='Backward', c='b')
    plt.plot(li_t, li_real_y, label='Real', c='k')
    plt.title('Solution with nt = %d, h = %5.4f' % (nt, h))
    plt.xlim(0,10) ; plt.ylim(-5,5)
    plt.xlabel('t') ; plt.ylabel('y')
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(8,6), facecolor='w')
    plt.plot(li_t, li_err_for, label='Forward', c='r')
    plt.plot(li_t, li_err_back, label='Backward', c='b')
    plt.title('Absolute error with nt = %d, h = %5.4f' % (nt, h))
    plt.xlim(0,10) ; plt.ylim(-2,2)
    plt.xlabel('t') ; plt.ylabel('y')
    plt.legend()
    plt.show()

    df = pd.DataFrame(data = {'Real': li_real_y,
                          'Forward': li_for_y, 
                          'Backward': li_back_y}, index=li_t)
    df['Error_forward'] = df['Forward'] - df['Real']
    df['Error_backward'] = df['Backward'] - df['Real']
    print(df)
```
