```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_start = 0 ; y_start = 1
t_end = 10
li_h = [1, 0.5, 0.25, 0.125]

plt.figure(figsize=(8,36), facecolor='w')

for i, h in zip(list(range(6)), li_h):
    li_t = [t_start]
    real_y = [y_start] ; for_y = [y_start] ; back_y = [y_start] ; tra_y = [y_start]
    
    t = t_start ; y = y_start
    while (t <= t_end):
        t += h
        li_t.append(t)
        real_y.append(np.exp(-t))
        
    t = t_start ; y = y_start
    while (t <= t_end):
        t += h
        y = (1-h/2)/(1+h/2)*y
        tra_y.append(y)
        
    t = t_start ; y = y_start
    while (t <= t_end):
        t += h
        y = 1/(1+h)*y
        back_y.append(y)
    
    t = t_start ; y = y_start
    while (t <= t_end):
        t += h
        y = (1-h)*y
        for_y.append(y)
        
    plt.subplot(6, 1, i+1)
    plt.plot(li_t, for_y, label='Forward')
    plt.plot(li_t, tra_y, label='Trapezoidal')
    plt.plot(li_t, back_y, label='Backward')
    plt.plot(li_t, real_y, label='Real')
    plt.title('h = %4.3f' % h)
    plt.xlim(0,)
    plt.legend()

plt.show()
```
